### USER INTERFACE CREATE

```python
import gradio as gr
def predict_unit_price(product_category_name, qty, total_price, freight_price, product_name_lenght,
                       product_description_lenght, product_weight_g, weekend, volume,
                       comp_1, comp_2, comp_3, lag_price):
    
    input_data = np.array([
        product_category_name, qty, total_price, freight_price, product_name_lenght,
        product_description_lenght, product_weight_g, weekend, volume,
        comp_1, comp_2, comp_3, lag_price
    ]).reshape(1, -1)

    # Apply scaling same used model training
    input_scaled = scaler.transform(input_data)
    
    # Predict with the scaled input
    prediction = model.predict(input_scaled)

    return f"Predicted Unit Price: ${prediction[0]:.2f}"
# "create a gradioo interface
iface=gr.Interface(
    fn=predict_unit_price,
    inputs=[
        gr.Number(label="Product Category Name"),
        gr.Number(label="Quantity"),
        gr.Number(label="Total Price"),
        gr.Number(label="Freight Price"),
        gr.Number(label="Product Name Length"),
        gr.Number(label="Product Description Length"),
        gr.Number(label="Product Weight (g)"),
        gr.Number(label="Weekend"),
        gr.Number(label="Volume"),
        gr.Number(label="Competitor Price 1"),
        gr.Number(label="Competitor Price 2"),
        gr.Number(label="Competitor Price 3"),
        gr.Number(label="Lag Price")
    ],
    outputs="text",
    title="Retail Unit Price Prediction",
    description="Enter the product and transaction details to predict the unit price."
)
# launch the interface 
iface.launch()
```
